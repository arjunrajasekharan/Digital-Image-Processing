```python
import numpy as np
import cv2
import math
from matplotlib import pyplot as plt
%matplotlib inline
img_full=cv2.imread('hdraw.png',0)
#Note orginal image resized to half as compuation time for the orginal one was too long
img_full = cv2.resize(img_full, (int(img_full.shape[1]*0.5),int(img_full.shape[0]*0.5 ))) 
plt.imshow(img_full,cmap="gray")
```

### Note : Ordinary correlation was found to produce poor results ,so the varient using mean was used

```python
def findbound(img_grey,subimg):
    img_cpy=img_grey.copy()
    out=np.zeros([(img_grey.shape[0]-subimg.shape[0]+1),(img_grey.shape[1]-subimg.shape[1]+1)],dtype = 'int64')
    values=[]
    val=0
    for i in range(0,subimg.shape[0]):
        for j in range(0,subimg.shape[1]):
            val+=subimg[i,j]
    subimg_mean=val/(subimg.shape[1]*subimg.shape[0])

    for x in range(0,(img_grey.shape[0]-subimg.shape[0])):
        for y in range(0,(img_grey.shape[1]-subimg.shape[1])):
            val=0
            for i in range(0,subimg.shape[0]):
                for j in range(0,subimg.shape[1]):
                    val+=img_grey[x+i,y+j]
                    tempimg_mean=val/(subimg.shape[1]*subimg.shape[0])
                        
            for i in range(0,subimg.shape[0]):
                for j in range(0,subimg.shape[1]):
                    out[x,y]+=int((img_grey[x+i,y+j]-tempimg_mean)*(subimg[i,j]-subimg_mean))
            values.append(out[x,y])
            #print(out[x,y])

    for i in range(0,out.shape[0]):
        for j in range(0,out.shape[1]):
            if(out[i,j]==(max(values))):
                c1=i
                c2=j
    print("Max value :",end="")
    print(max(values))
    print("Top left corner : ",end="")
    print(c1,c2)

#Drawing boundary
    for i in range(c1,c1+subimg.shape[0]):
            img_cpy[i,c2]=255
            img_cpy[i,c2+subimg.shape[1]]=255
    for j in range(c2,c2+subimg.shape[1]):
        img_cpy[c1,j]=255
        img_cpy[c1+subimg.shape[0],j]=255
    plt.imshow(img_cpy,cmap='gray')
    plt.show() 
```
